The dataset is not as balanced as initially thought. It may be that there are not as many observations in the folders or it may be that errors occurred while communicating with Google to retrieve the vectors. Another possibility is that there were issues with the processing. Either way, there may be difficulty signing nothing (n) because of only 6 observations compared to the other gestures which have over 2000 examples.

One approach is to drop the (n) as there are too few samples, and undersample the other letters so that they have the same quantity of observations of the smallest sample. This would be N at 1417 observations.

## Undersampling the observations

The observations are being undersamples so they are more uniform and to reduce the influence of any enoromously large quantity of observations as those in F.

```python
# Remove the nothing (n) observations

working_df = df.drop(df[df.Label =='n'].index, axis=0)

# Get the unique labels.
labels = working_df['Label'].unique()
```

```python
labels
```

```python
import numpy as np

balanced_ids = working_df[working_df['Label'] == 'N'].index
limit = len(balanced_ids)
for l in labels:
  x_ids = working_df[working_df['Label'] == l].index
  if l != 'N':
    balanced_ids = balanced_ids.union(np.random.choice(a=x_ids, size=limit))

# The number of labels should be 28 x 1417 = 39676
len(balanced_ids)
```

```python
balanced_df = df.loc[balanced_ids]
```

```python
val_counts = balanced_df['Label'].value_counts()
val_counts
```

```python
val_counts.plot.bar()
```
